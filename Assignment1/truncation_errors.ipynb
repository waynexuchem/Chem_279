```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("truncation_errors.csv", sep="\t", header=None)
data.columns = ["h_values", "forward_errors", "central_errors"]
data
```

```python
plt.scatter(np.log(abs(data["h_values"])), np.log(abs(data["forward_errors"])), label="Forward Errors")
plt.scatter(np.log(abs(data["h_values"])), np.log(abs(data["central_errors"])), label="Central Errors")

plt.xlabel("Log(h_values)")
plt.ylabel("Log(errors)")
plt.title("Log-Log Plot of h_values vs. Errors")
plt.legend()

plt.savefig("Log-Log Plot of h_values vs. Errors.png", dpi=600)

plt.show()
```
